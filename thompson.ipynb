```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("https://raw.githubusercontent.com/Komal-99/Reinforcement-Learning/main/Ads_Optimisation.csv")
```

```python
data
```

```python
import random
N=data.shape[0]
d=data.shape[1]
ads_selected=[]
number_of_rewards_1=[0]*d
number_of_rewards_0=[0]*d
total_reward=0
for n in range(N):
  ad=0
  max_random_beta=0
  for i in range(d):
    random_beta=random.betavariate(number_of_rewards_1[i]+1,number_of_rewards_0[i]+1) 
    if random_beta>max_random_beta:
      max_random_bet=random_beta
      ad=i
    ads_selected.append(ad)
    reward=data.iloc[n,ad]
    if reward==1:
      number_of_rewards_1[ad]+=1
    else:
      number_of_rewards_0[ad]+=1
    total_reward +=reward
```
