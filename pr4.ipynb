```python
import pandas as pd 
df=pd.DataFrame({'a':np.random.randint(0,50,1000)})
df['b']=df['a']+np.random.normal(0,10,1000) #possitive correlation
df['c']=df['a'] + np.random.normal(0,5,1000) # negative coreelation
df['d']= np.random.randint(0,50,1000) # not correlated with'a'
```
